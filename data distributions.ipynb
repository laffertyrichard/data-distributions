```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gammadist = np.random.gamma(10, 10, 100)
```

```python
plt.hist(gammadist)
plt.axvline(gammadist.mean(), color = 'red', linestyle = 'solid')
plt.axvline(gammadist.mean() + gammadist.std(), color = 'red', linestyle = 'dashed')
plt.axvline(gammadist.mean() - gammadist.std(), color = 'red', linestyle = 'dashed')
```

Data is skewed right which is what one would expect with a gamma distribution

```python
normal_1 = np.random.normal(5, .5, 100)
normal_2 = np.random.normal(10, 1, 100)
normal_3 = normal_1 + normal_2
```

```python
plt.hist(normal_3)
plt.axvline(normal_3.mean(), linestyle = 'solid', color = 'black' )
plt.axvline(normal_3.mean() + normal_3.std(), linestyle = 'solid', color = 'coral' )
plt.axvline(normal_3.mean() - normal_3.std(), linestyle = 'solid', color = 'coral' )
```

It appears that the data is slightly skewed right.

```python
plt.plot(normal_1, gammadist, 'o')
```
